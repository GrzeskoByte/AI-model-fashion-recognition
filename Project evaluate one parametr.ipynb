```python
import tensorflow as tf
import numpy as np
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img,img_to_array
```

```python
def image_to_array(path):
    img_array = []

    image = plt.imread(path)
    img_array.append(image.astype('float32')/255)
    
    return np.array(img_array)
    

def get_key_by_value(dictionary, value):
    for key, val in dictionary.items():
        if val == value:
            return key
    return None

def preprocess_image(image):
    target_size = (60, 80)
    
    if isinstance(image, str):
        img = Image.open(image)
    else:
        img = image
    

    img = img.resize(target_size)
    img_array = np.array(img)

    img_array = img_array / 255.0
    
    img_array = np.expand_dims(img_array, axis=0)
    
    return img_array


image_path = os.getcwd() + r"\archive\images\11373.jpg"
prepared_image =image_to_array(image_path)

sunglasses_image_path = os.getcwd() + r"\archive\images\21279.jpg"
prepared_sunglasses = image_to_array(sunglasses_image_path)

sport_shoes_image_path= os.getcwd() + r"\archive\images\5465.jpg"
prepared_shoes = image_to_array(sport_shoes_image_path)

bottomwear_image_path = os.getcwd() + r"\archive\images\14599.jpg"
prepared_bottomwear = image_to_array(bottomwear_image_path)

nastka_path = os.getcwd() + r"\test.jpg"
nastka_image = preprocess_image(nastka_path)

kuba_path =os.getcwd() + r"\kuba.jpg"
kuba_image = preprocess_image(kuba_path)

hat_path = os.getcwd() + r"\czapka.jpg"
hat_image = preprocess_image(hat_path)
```
